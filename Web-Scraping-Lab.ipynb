## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Import the required libraries
import requests
from bs4 import BeautifulSoup
import csv
```

Download the webpage at the url


```python
# The URL containing the data to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"

# Download the webpage
data = requests.get(url).text
```

Create a soup object


```python
# Create a BeautifulSoup object
soup = BeautifulSoup(data, 'html.parser')
```

Scrape the `Language name` and `annual average salary`.


```python
# Scrape the Language name and annual average salary
# Find the table by its tag (assuming it's the only one or easy to identify)
table = soup.find('table')

# Create a list to store the scraped data (Language, Annual Average Salary)
scraped_data = []

# Iterate through all table rows (tr)
for row in table.find_all('tr'):
    # Get all columns (td) in that row
    cols = row.find_all('td')
    
    # Check if there are enough columns and skip the header row (which has 'th' or no data)
    if len(cols) > 0:
        # The programming language name is typically in the first column (index 1 in the table, but index 0 in the cols list)
        language_name = cols[1].getText()
        # The annual average salary is typically in the fourth column
        annual_salary = cols[3].getText()
        
        # Append the extracted data to the list
        scraped_data.append([language_name, annual_salary])

# Optional: Print the list to verify the scraped data
print(scraped_data)
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# Define the name of the CSV file
csv_file_name = 'popular-languages.csv'

# Open the CSV file in write mode ('w')
with open(csv_file_name, 'w', newline='') as file:
    # Create a CSV writer object
    writer = csv.writer(file)
    
    # Write the header row
    writer.writerow(['Language', 'Annual Average Salary'])
    
    # Write the scraped data rows
    writer.writerows(scraped_data)

print(f"Data successfully saved to {csv_file_name}")
```
